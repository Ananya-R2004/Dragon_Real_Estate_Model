###  Objective

To split the dataset into **training (80%)** and **testing (20%)** sets while preserving the **distribution of the `CHAS` feature** (Charles River dummy variable).

---

###  Why 80-20 Split?

- **80% Training Set**: Used by the model to learn patterns.
- **20% Test Set**: Used to evaluate how well the model performs on unseen data.
- Prevents **overfitting** and ensures **realistic model evaluation**.

---

###  Why Stratify on `CHAS`?

- `CHAS` is a binary feature (`0` = not near river, `1` = near river).
- It’s **imbalanced** (e.g., only ~7% are `CHAS = 1`).
- **Random split** may miss the minority class in the test set.
- Stratified sampling keeps the **same class distribution** in both train and test sets.

# Reading & Interpreting Pair-Plot Matrix

A pair-plot (scatter-histogram matrix) is a powerful way to **visualize distributions** and **pairwise relationships** at once. Below is a unified guide to:

1. **How to read & analyze** any pair-plot, step by step.  
2. **Concrete insights** drawn from 4×4 matrix of **MEDV**, **ZN**, **RM**, and **LSTAT**.

---

## 1. Diagonals: Marginal Distributions

Each diagonal cell is a **histogram** for one variable. Use it to assess:

| Question            | Example (Your Plot)                                                 |
|---------------------|----------------------------------------------------------------------|
| **Shape**           | - **MEDV**: Bell-shaped 10–35 k + spike at 50 k cap   |
|                     | - **ZN**: Zero-inflated (many zeros)                                 |
|                     | - **RM**: Roughly symmetric around ~6 rooms                         |
|                     | - **LSTAT**: Right-skewed, tail out to ~38%                         |
| **Spread & Range**  | See MEDV bars from 5–50 k; ZN mostly 0 with a few at 12.5 & 100     |
| **Outliers & Gaps** | Notice the solitary high-value bin at MEDV=50 (capped data).        |

> **Usage:** Identify skewness for log/√ transforms, caps to flag, or bins for zero-inflated features.

---

## 2. Off-Diagonals: Pairwise Scatter Cells

Each off-diagonal subplot at row _i_, column _j_ shows **Feature _i_ (y) vs. Feature _j_ (x)**. Follow this mini-checklist:

| Checklist             | Example (Your Plot)                                                                                           |
|-----------------------|---------------------------------------------------------------------------------------------------------------|
| **Trend**             | - **MEDV vs RM**: Clear upward slope → positive association                                                   |
|                       | - **MEDV vs LSTAT**: Clear downward slope → negative association                                              |
| **Tightness**         | - **MEDV vs RM**: Tightly clustered → strong correlation                                                     |
|                       | - **MEDV vs ZN**: Wide cloud → weak correlation                                                               |
| **Linearity vs Curve**| - **MEDV vs LSTAT**: Nearly linear downward                                                               |
|                       | - **ZN vs RM**: No clear line (no strong functional form)                                                     |
| **Outliers**          | - **MEDV vs ZN**: A few high-ZN points with variable MEDV → investigate zoning anomalies                     |
| **Density**           | - High point density for RM between 5.5–7 and MEDV 15–30 k                                                    |

> **Tip:** Always read the x-axis label at the bottom and y-axis label on the left.

---

## 3. Cross-Comparison & Symmetry

- The cell at _(RM, LSTAT)_ mirrors _(LSTAT, RM)_—axes swap but pattern is identical (downward cloud).
- If a feature is skewed on the diagonal (e.g., ZN), its scatterplots may bunch points at zero—consider binning.

#  ML Pipeline Summary: Imputer + Scaler

This pipeline performs two core data preprocessing steps on numerical features before feeding data to a machine learning model.

---

##  Pipeline Components

### 1. ("imputer", SimpleImputer(strategy="median"))

- **Purpose**: Handles missing values.
- **Strategy**: Replaces missing values with the **median** of each column.
- **Why Median?**: Robust to outliers and skewed distributions.

 *Useful when:*
- Your data has missing values.
- You want to avoid the influence of extreme values (outliers).

---

### 2. ("std_scaler", StandardScaler())

- **Purpose**: Scales features to a standard normal distribution.
- **Effect**: Transforms each feature to have:
  - Mean = 0
  - Standard deviation = 1
- **Why Needed?**
  - Some models (like SVM, KNN, linear regression) perform better when all features are on the same scale.

---
